```python
import numpy as np
from random import randint
from matplotlib import pyplot as plt
from math import exp
%matplotlib inline
```

```python
num_iters = 100000
num_tosses = 10
num_coins = 1000
v_1 = []
v_min = []
v_ran = []
for test in range(num_iters):
    flips = np.random.randint(0, 2, (num_coins, num_tosses))
    fraction_of_head = flips.sum(axis=1)
    v_1.append(fraction_of_head[0])
    v_min.append(min(fraction_of_head))
    v_ran.append(fraction_of_head[randint(0, 999)])
```

```python
print(np.mean(v_1))
print(np.mean(v_ran))
print(np.mean(v_min))
```

```python
plt.xlabel('number of heads')
plt.ylabel('frequency')
plt.title('v_1')
plt.hist(v_1, bins=10)
```

```python
plt.xlabel('number of heads')
plt.ylabel('frequency')
plt.title('v_ran')
plt.hist(v_ran, bins=10)
```

```python
plt.xlabel('number of heads')
plt.ylabel('frequency')
plt.title('v_min')
plt.hist(v_min, bins=10)
```
